Q1. Explain why we have to use the Exception class while creating a Custom Exception.
Note: Here Exception class refers to the base class for all the exceptions.

Ans: All exceptions in Python inherit from the Exception base class. If we don’t inherit from it, our custom exception won’t behave like a standard Python error and won’t be caught properly.

```python
class MyError(Exception):
    pass

try:
    raise MyError("Custom exception raised")
except MyError as e:
    print("Caught:", e)
```

Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

Ans: ArithmeticError is the base class for numeric calculation errors:- ZeroDivisionError → raised when dividing by zero, OverflowError → raised when a calculation exceeds the limit.

```python
# ZeroDivisionError
try:
    10/0
except ZeroDivisionError as e:
    print(e)

# OverflowError
import math
try:
    math.exp(1000)
except OverflowError as e:
    print(e)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans: LookupError is the base class for invalid key/index errors:- KeyError → raised when a dictionary key is missing, IndexError → raised when a list index is out of range.

```python
# KeyError
try:
    d = {"a":1}
    print(d["b"])
except KeyError as e:
    print(e)

# IndexError
try:
    lst = [1,2,3]
    print(lst[5])
except IndexError as e:
    print(e)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

Ans: ImportError → raised when importing fails (function/class not found), ModuleNotFoundError → subclass of ImportError, raised when the module itself doesn’t exist.

```python
# ImportError
try:
    from math import cube
except ImportError as e:
    print(e)

# ModuleNotFoundError
try:
    import abcxyz
except ModuleNotFoundError as e:
    print(e)
```

Q6. List down some best practices for exception handling in python.

Always inherit custom exceptions from Exception, Catch specific exceptions, not broad except:, Use finally for cleanup tasks, Avoid silent exception handling, Use logging instead of only print, Provide meaningful error messages.
